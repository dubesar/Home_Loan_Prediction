```
import numpy as np
import pandas as pd
x=pd.read_csv('train_u6lujuX_CVtuZ9i.csv')
y=pd.read_csv('test_Y3wMUE5_7gLdaTN.csv')
```

```
z=x['Loan_Status']
```

```
x=x.drop(['Loan_Status'],axis=1)
```

```
x=x.append(y,ignore_index=True)
```

```
x
```

```
x['Gender']=x['Gender'].replace(['Male','Female','NaN'],[1,0,2])
```

```
x['Gender'].fillna(0,inplace=True)
```

```
x['Married']=x['Married'].replace(['No','Yes'],[0,1])
x['Education']=x['Education'].replace(['Graduate','Not Graduate'],[1,0])
x['Self_Employed']=x['Self_Employed'].replace(['Yes','No'],[1,0])
x['Property_Area']=x['Property_Area'].replace(['Urban','Semiurban','Rural'],[0,1,2])
x['Dependents']=x['Dependents'].replace(['3+'],3)
```

```
x['Dependents'].fillna(0,inplace=True)
x['Self_Employed'].fillna(0,inplace=True)
x['LoanAmount'].fillna(x['LoanAmount'].mean(),inplace=True)
x['Credit_History'].fillna(0,inplace=True)
```

```
x
```

```
z
```

```
z=z.replace(['Y','N'],[1,0])
```

```
z
```

```
from sklearn import *
```

```
from sklearn.model_selection import train_test_split
```

```
full_X=x.iloc[:,1:]
```

```
full_X=full_X.iloc[:,:]
```

```
full_X['Loan_Amount_Term'].fillna(360,inplace=True)
full_X.fillna(0)
```

```
full_X
full_X['Married'].fillna(0,inplace=True)
```

```
train_valid_X = full_X[ 0:614 ]
train_valid_y = z
test_X = full_X[ 614: ]
train_X , valid_X , train_y , valid_y = train_test_split( train_valid_X , train_valid_y , train_size = .7 )

print (full_X.shape , train_X.shape , valid_X.shape , train_y.shape , valid_y.shape , test_X.shape)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import metrics
model = LogisticRegression()
```

```
train_X.fillna(0)
train_y.fillna(0)
```

```
train_X.isnull().any()
```

```
train_y.isnull().any()
```

```
train_X['Married'].fillna(0)
```

```
model.fit(train_X,train_y)
```

```
print (model.score( train_X , train_y ) , model.score( valid_X , valid_y ))
```

```
test_Y = model.predict(test_X).astype(int)
test_Y=np.where(test_Y==1,"Y","N")
Loan_ID = x[614:].Loan_ID
test = pd.DataFrame( { 'Loan_ID': Loan_ID , 'Loan_Status': test_Y } )
test.shape
test.head()
test.to_csv( 'test_csv.csv' , index = False )
```
